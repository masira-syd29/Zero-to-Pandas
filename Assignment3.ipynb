```python
file_path = './covid-countries-only.csv'
covid_data_df = pd.read_csv(file_path)
covid_data_df
```

Q. Count the number of countries for which the total test data is missing

```python
total_tests_missing = covid_data_df['total_tests'].isna().sum()
print(f'The data for total tests is missing for {total_tests_missing} countries')
```

Q. Merge country_df and covid_data_df on the location column

```python
combined_df = pd.merge(left=country_df, right=covid_data_df, on='location')
combined_df
```

```python
merged_df = pd.merge(country_df, covid_data_df, on='location', how="inner")
merged_df
```

Q. Add columns tests_per_million, cases_per_million, deaths_per_million into combined_df

```python
combined_df['tests_per_million'] = combined_df['total_tests'] *1e6 / combined_df['population']
combined_df['cases_per_million'] = combined_df['total_cases'] * 1e6 / combined_df['population']
combined_df['deaths_per_million'] = combined_df['total_deaths'] * 1e6 / combined_df['population']
combined_df
```

Q. Create a dataframe with 10 countries that have the highest number of tests per million

```python
highest_tests_df = combined_df.sort_values(by='tests_per_million', ascending=False).head(10)
highest_tests_df
```

Q. Create a dataframe with 10 countries that have the highest number of positive cases per million

```python
highest_cases_df = combined_df.sort_values(by='cases_per_million', ascending=False).head(10)
highest_cases_df
```

Q. Create a dataframe with 10 countries that have the highest number of death cases per million

```python
highest_deaths_df = combined_df.sort_values(by='deaths_per_million', ascending=False).head(10)
highest_deaths_df
```

Q. Count the number of countries that feature in both the lists of 'highest number of tests per million' and 'highest number of cases per million'

```python
set_tests = set(highest_tests_df['location'].tolist())
set_tests
```

```python
set_cases = set(highest_cases_df['location'].tolist())
set_cases
```

```python
common_countries = set_tests.intersection(set_cases)
common_countries
```

```python
count_common_countries = len(common_countries)
print(f"The number of countries that feature in both lists is: {count_common_countries}")
```

  Q. Count the number of countries that feature in both the lists '20 countries with lowest GDP per capita' and '20 countries with lowest number of hospital beds per thousand population'. Only consider countries with a population higher than 10 million while creating the list

```python
countries_with_hundred_million_population = combined_df[combined_df['population'] > 10000000]
countries_with_hundred_million_population
```

```python
lowest_gdp_countries = countries_with_hundred_million_population.sort_values(by='gdp_per_capita', ascending=True).head(20)['location'].tolist()
lowest_gdp_countries
```

```python
lowest_beds_countries = countries_with_hundred_million_population.sort_values(by='hospital_beds_per_thousand', ascending=True).head(20)['location'].tolist()
lowest_beds_countries
```

```python
same_countries = set(lowest_gdp_countries).intersection(set(lowest_beds_countries))
same_countries
```

```python
count_same_countries = len(same_countries)
print(f"The number of countries that feature in both lists is: {count_same_countries}")
```
